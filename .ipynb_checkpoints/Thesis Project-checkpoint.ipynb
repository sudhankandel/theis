```python
from glob import glob
import os
import pandas as pd
from nltk.corpus import stopwords
import re
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import preprocessing


from keras.preprocessing.sequence import pad_sequences
from keras.utils.np_utils import to_categorical
from keras.preprocessing.text import Tokenizer
```

```python
arr = os.listdir('dataset')
raw_data=[]
category=[]
for i in arr:
    file_list = glob('dataset/'+i+"/*.txt")
    for file_path in file_list:
        try:
            with open(file_path,encoding='utf8') as f:
                raw_data.append(f.read())
                category.append(i)
        except:
            pass
```

```python
for i in arr:
    file_list = glob('dataset/'+i+"/*.txt")
    for file_path in file_list:
        try:
            with open(file_path,encoding='utf16') as f:
                raw_data.append(f.read())
                category.append(i)
        except:
            pass        
```

```python
df_basic={'label':category,
         'news':raw_data}
raw_df=pd.DataFrame(df_basic)
raw_df     
```

```python
stop = stopwords.words('nepali')
raw_df['stopwords'] = raw_df['news'].apply(lambda x: len([x for x in x.split() if x in stop]))
raw_df[['news','stopwords']].head()
```

```python
stop_words = stopwords.words("nepali")
def string_manipulation(df,column)  : 
    df[column] = df[column].apply(lambda x: re.sub('[।(),०-९<<?!,—,–,/,’,‘,:,\u200d,\ufeff]', '', x))
    df[column] = df[column].apply(lambda x: " ".join([i for i in x.split()if i not in (stop_words)]))

    return df
Data = string_manipulation(raw_df,"news")
Data = Data.sample(frac=1,random_state=20).reset_index(drop=True)
Data
```

```python
plt.figure(figsize = (12,12))

ax = sns.countplot(y = Data['label'] ,
                   linewidth = 1,edgecolor = "k"*2,
                   palette = "Reds_r")

for i,j in enumerate(Data['label'].value_counts().values) :
    ax.text(0.7,i,j,fontsize = 15)

plt.grid(True)
plt.ylabel("")
plt.title("Total category")
plt.show()
```
